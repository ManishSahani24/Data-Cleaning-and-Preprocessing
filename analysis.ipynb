#### 2. Handling missing values in "Age" :

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the specific CSV file ---
df = pd.read_csv('Titanic-Dataset.csv')

# Calculate the median from YOUR data's 'Age' column.
# The .median() function automatically handles and ignores missing values.
median_age = df['Age'].median()

print(f"The exact median age from your file is: {median_age}")

# --- 3. Plot the data ---
df_for_plotting = df.copy() # Create a copy to avoid changing original data

# Create the histogram and median line
plt.figure(figsize=(10, 6))
plt.hist(df_for_plotting['Age'], bins=30, color='teal', edgecolor='black', alpha=0.8, label='Age Distribution')
plt.axvline(median_age, color='orange', linestyle='dashed', linewidth=2, label=f'Median Age: {median_age:.2f}')

# Add titles and labels
plt.title('Distribution of Ages from Your Dataset', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the plot to a file
plt.savefig('your_data_age_median_plot.png')
```

#### We are going to replace the missing entry of "age" with the median value of age and save it in a new csv file - "dataset_cleaned.csv"

```python
import pandas as pd

# 1. Load the original, untouched data
df = pd.read_csv('Titanic-Dataset.csv')

# 2. Calculate the value to be used for imputation
median_age = df['Age'].median()

# 3. Apply the change using code
df['Age'] = df['Age'].fillna(median_age)

# 4. Save the result to a new file
df.to_csv('dataset_cleaned.csv', index=False)

print("Data cleaning complete. Results saved 'dataset_cleaned.csv' .")
```

#### 4.  Handling missing value in Embarked: 

    We fill the missing values with the most common port of embarkation (the mode).

```python
import seaborn as sns
# --- 1. Find the mode ---

mode_port = df['Embarked'].mode()[0]
print(f"The mode (most frequent port) is: '{mode_port}'")


# --- 2. Plot the frequency of each port ---

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=df, order=df['Embarked'].value_counts().index)

# Add a title and labels
plt.title(f"Frequency of Embarkation Ports (Mode is '{mode_port}')", fontsize=14)
plt.xlabel('Port of Embarkation', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)

# Save the plot to a file
plt.savefig('embark_mode_plot.png')
```
